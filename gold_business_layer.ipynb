```sql
create or replace table gold_category_performance
as
select count(distinct sp.product_id) as liczba_produktow_w_kategorii,
       avg(sp.price) as srednia_cena_w_kategorii,
       sum(sp.price*sp.stock) as wartosc_w_kategorii
from silver_products sp
group by sp.category
order by wartosc_w_kategorii desc

```

```sql
create or replace table gold_price_segment_stats as
with cte as (
select *, 
        case when price < 20 then 'cheap'
        when price between 20 and 50 then 'medium'
        else 'expensive' end as price_category
from silver_products

)
select price_category,avg(rating) as sredni_rating_w_segmencie, avg(discountPercentage) as sredna_znizka_w_segemncie from cte
group by price_category
```
